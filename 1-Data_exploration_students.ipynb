```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Selecting only numeric columns
numeric_df = df.select_dtypes(include='number')

# Computing correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Student Scores")
plt.show()
```

```python
#Heat-map Observations
1. Strong positive correlation between Reading and Writing scores ( 0.95)
This indicates that students who score high in reading tend to also score high in writing.
The relationship is very strong, suggesting closely related skills.

2. Strong positive correlation between Math and Reading scores ( 0.82)
Students with higher math scores generally perform well in reading.
However, the relationship is slightly weaker than readingwriting.
    
3. Strong positive correlation between Math and Writing scores ( 0.80)
Math and writing scores also increase together, but this is the weakest correlation among the three pairs.
This suggests math performance is somewhat less related to writing compared to reading.

4. Diagonal values are 1
This represents perfect correlation of each subject with itself, which is expected in a correlation matrix.

5. Overall insight
All three subjects are positively correlated, meaning improvement in one subject is generally associated with improvement in the others.
```

```python
#Scatter-Plot to visualize relationship between Math and Reading Scores
plt.figure(figsize=(7,5))
sns.scatterplot(x="math score", y="reading score", data=df)  # replace column names if needed
plt.title("Reading vs Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
#Scatter-Plot Observation
1. Clear positive linear trend
As math scores increase, reading scores also tend to increase.
This visually supports the correlation value (~0.82) seen in the heatmap.

2. Most data points are concentrated in the mid-to-high score range
A large number of students score between roughly 5080 in both math and reading.
    
3. Some dispersion around the trend line
While the trend is strong, the points are not perfectly aligned, indicating individual variation in student performance.

4. Very few extreme outliers
Most students follow the general pattern, suggesting consistent scoring behavior across subjects.
```
